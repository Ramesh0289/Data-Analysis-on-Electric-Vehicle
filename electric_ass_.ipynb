```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# univariate analysis
```

```python
sns.histplot(df['Model Year'], kde=True)
plt.title('Distribution of Model Year')
plt.xlabel('Model Year')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(df['Model Year'])
plt.title('Distribution of Model Year')
plt.xlabel('Model Year')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['Electric Range'], kde=True)
plt.title('Distribution of Electric Range')
plt.xlabel('Electric Range')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x=df['Electric Range'])
plt.title('Box Plot of Electric Range')
plt.xlabel('Electric Range')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(x='Make', data=df)
plt.title('Count of Vehicles by Make')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x='County', data=df)
plt.title('Distribution of County Category')
plt.xlabel('County Category')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x='Electric Vehicle Type', data=df)
plt.title('Electric Vehicle Type')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
#Bivariate analysis
```

```python
#categorical data vs Numerical data
sns.boxplot(x='Electric Vehicle Type', y='Electric Range', data=df)
plt.title('Electric Range Distribution by Electric Vehicle Type')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Electric Range')
plt.show()
```

```python
#Numerical data vs Numerical data
sns.scatterplot(x='Model Year', y='DOL Vehicle ID', data=df)
plt.title('Model Year vs DOL Vehicle ID')
plt.xlabel('Model Year')
plt.ylabel('DOL Vehicle ID')
plt.show()
```

```python
sns.scatterplot(x='Model Year', y='Electric Range', data=df)
plt.title('Model Year vs Electric Range')
plt.xlabel('Model Year')
plt.ylabel('Electric Range')
plt.show()
```

```python
#categorical data vs categorical data
plt.figure(figsize=(12, 6))
sns.countplot(x='Make', hue='Electric Vehicle Type', data=df)
plt.title('Electric Vehicle Type vs Make')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Electric Vehicle Type')
plt.show()
```

```python
import plotly.express as pl
tem_df = df.copy(deep=True)
tem_df.drop(tem_df.columns[[0,8,9,11,12,13,15,16]],axis=1,inplace=True)
fig = pl.choropleth(
    tem_df,
    locations = 'State',
    color = "Make",
    hover_name= "County",
    hover_data= ["Electric Range","City","State"],
    locationmode='USA-states',
    animation_frame = "Model Year",
    scope="usa",
)
fig.update_geos(overwrite=True)
fig.show()
```
